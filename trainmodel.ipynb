```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import pickle

```

```python

# Step 1: Load your dataset
# Replace with your dataset path
data = pd.read_csv('Admission_Prediction.csv')
data
```

```python
data.isna().sum()
```

```python
data
```

```python
data.drop(columns="Serial No.",inplace=True)
```

```python
data['GRE Score'].fillna(data['GRE Score'].mean(), inplace=True)
data['TOEFL Score'].fillna(data['TOEFL Score'].mean(), inplace=True)
data['University Rating'].fillna(data['University Rating'].mean(), inplace=True)
```

```python

# Step 2: Prepare input features (X) and target (y)
# Replace these with your actual column names
X = data[['GRE Score', 'TOEFL Score', 'University Rating',
          'SOP', 'LOR', 'CGPA', 'Research']]
y = data['Chance of Admit']  # Replace with your target variable column
```

```python
# Step 3: Initialize and train the model
model = LinearRegression()
model.fit(X, y)

# Step 4: Save the model to a pickle file
with open('finalized_model1.pickle', 'wb') as f:
    pickle.dump(model, f)

print("Model trained and saved successfully.")
```
